```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score
from sklearn.metrics import silhouette_score
from sklearn import preprocessing
```

```python
wine_data = pd.read_csv('wine.data',header=None)
wine_data_arr = np.array(wine_data)
m,n = wine_data_arr.shape
wine_label = wine_data_arr[:,0].astype(int)
print("实验原始数据： ")
print(wine_label)
scaler = preprocessing.MinMaxScaler()
wine_data_scaled = scaler.fit_transform(wine_data_arr[:,1:-1])
y_pred = KMeans(n_clusters=3,random_state=10).fit(wine_data_scaled)
y_pred.labels_ += 1
print("实验聚类结果： ")
print(y_pred.labels_)
f1_micro = f1_score(wine_label,y_pred.labels_,average='micro')
f1_macro = f1_score(wine_label,y_pred.labels_,average='macro')
sil_sco = silhouette_score(wine_data_scaled,y_pred.labels_)
print(f1_micro)
print(f1_macro)
print(sil_sco)
```

```python
import sklearn.cluster as skc
db = skc.DBSCAN(eps=0.49,min_samples=4).fit(wine_data_scaled)
labels = db.labels_+2
print("实验原始数据： ")
print(wine_label)
print("实验预测结果： ")
print(labels)
f1_micro_db = f1_score(wine_label,labels,average='micro')
f1_macro_db = f1_score(wine_label,labels,average='macro')
sil_sco_db = silhouette_score(wine_data_scaled,labels)
print(f1_micro_db)
print(f1_macro_db)
print(sil_sco_db)
```
